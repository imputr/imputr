```python
import pandas as pd
import numpy as np
from imputr import AutoImputer

df = pd.read_csv('datasets/DigiDB_digimonlist.csv')

# Create random missing values
np.random.seed(100)
mask = np.random.choice([True, False], size=df.shape)
df.mask(mask, inplace=True)

imputer = AutoImputer(df)
imputed_df = imputer.impute()
```
